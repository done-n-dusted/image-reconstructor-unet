```python
import matplotlib.pyplot as plt
import cv2
import os
from PIL import Image
import torch
import torchvision
```

```python
img_src = '../data/raw/Airport/001.jpg'
```

```python
img = cv2.imread(img_src)
```

```python
h, w, c = img.shape
```

```python
plt.imshow(img)
```

```python
import numpy as np
```

```python
def create_mask(h, w, c):
    mask = np.zeros((h, w), dtype=np.uint8)
    blobs = 8

    for _ in range(blobs):
        cx = np.random.randint(0, w)
        cy = np.random.randint(0, h)
        r = np.random.randint(5, 30)
        cv2.circle(mask, (cx, cy), r, 255, -1)

    mask = np.stack([mask]*3, axis = -1)
    return mask
    
```

```python
mask = create_mask(h, w, c)
```

```python
res = np.maximum(img, mask)
```

```python
plt.imshow(np.concat([img, mask, res], axis = 1))
```
